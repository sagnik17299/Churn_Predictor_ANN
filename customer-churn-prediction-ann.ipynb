```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as skl
from sklearn import linear_model
import tensorflow
import keras
import mlxtend
```

```python
df = pd.read_csv('/kaggle/input/bank-customer-churn-dataset/Bank Customer Churn Prediction.csv')
```

```python
df
```

```python
df1 = df.drop(['customer_id'],axis=1)
```

```python
df1
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['country'] = le.fit_transform(df1['country'])
df1['gender'] = le.fit_transform(df1['gender'])
```

```python
x = df1.iloc[:,:-1]
y = df1['churn']
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x = pd.DataFrame(ss.fit_transform(x),columns = x.columns)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
from keras.layers import Dense
from keras.models import Sequential
from keras.callbacks import EarlyStopping 
```

```python
ann = Sequential()
```

```python
ann.add(Dense(8,input_dim= 10,activation='relu'))
ann.add(Dense(6,activation='relu'))
ann.add(Dense(4,activation='relu'))
ann.add(Dense(2,activation='relu'))
ann.add(Dense(1,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size=100,epochs=50,validation_data=(x_test,y_test),callbacks=EarlyStopping())
```

```python
y_pred1=ann.predict(x_train)
prd_data1 = []
for i in y_pred1:
    if i[0] > 0.5:
        prd_data1.append(1)
    else:
        prd_data1.append(0)
```

```python
y_pred=ann.predict(x_test)
prd_data = []
for i in y_pred:
    if i[0] > 0.5:
        prd_data.append(1)
    else:
        prd_data.append(0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prd_data)*100
```

```python
accuracy_score(y_train,prd_data1)*100
```
